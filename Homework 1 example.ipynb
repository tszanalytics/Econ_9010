```julia
# regression function
function linreg(y,x)
  # add intercept
  n = length(y)
  X = [ones(n) x]
  b = X \ y
  resids = y - X*b
    
  s2 = sum(resids.^2)/n
  covb = s2*inv(X'X)  
  seb = diag(covb)
  
  println("coeffs = ",round.(b,3))
  println(" s.e's = ",round.(seb,3))
  println("s^2 (eqn. variance) = ",round(s2,6))

  # compute R^2
  tss = sum((y .- mean(y)).^2)
  R2 = 1 - s2*n/tss
  println("Rsquared = ",round(R2,3))
  s = sqrt(s2)
  return b, seb, s, R2
end

b, s, R2 = linreg(y,x)
```

```julia
## using the formula:
function student_t(x,mu,sig,v)
    lc = lgamma((v+1)/2) - 0.5*log(pi*v*sig^2) - lgamma(v/2)
    k = (1 + ((x - mu)^2)/(v*sig^2))^(-(v+1)/2)
    return exp(lc)*k
end
# THIS WORKS - formula gives correct scale, but
# the function TDist in Distributions does not
# seem to - not sure why.
```
